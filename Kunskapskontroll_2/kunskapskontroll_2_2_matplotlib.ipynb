```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

It can be done either in the OO(object oriented style) or the pyplot style. The pyplot style provides an interface similar to working in matlab.
It is recommended to use the OO style for most graphs, especially those that are to be used in larger projects or for embedding in GUI's.
But the pyplot style can be "Convenient" for simple graphs nad "Quick interactive work".

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

A Figure is the main object that contains the graphical data for the plot. 

An Axes is an Artist attached to the Figure that contain a region for plotting data, 
it usually contains 2 Axis objects. You can also unse methods on it to set attributes such as labels and legends

An Axis contains and sets the scale of the plots, generates the ticks and ticklabels.

Almost everything in a Figure is an Artist, it is every object that can draw something on the canvas.

### When plotting in Matplotlib, what is the expected input data type?

The expected data type is numpy.ndarray or objects that can be passed to numpy using the .asarray method. And even if you can plot other types of data is is recommended to convert the data to numpy arrays before so the graphs will behave as intended

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0, *, device=None)
x = np.linspace(-4, 4, 100)  # Sample data.
y = x**2

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, y, label='Quadratic')  # Plot some data on the Axes.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("My OO Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.


plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, y, label='Quadratic')  # Plot some data on the (implicit) Axes.
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("My pyplot Plot")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)


# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), layout='constrained')
ax1.scatter(random_data_x, random_data_y, label='Scatter',s=5, color='g')  # Size = 5, color = green
ax2.bar(names, values, label='Fruit Data', color='g') 

ax1.set_xlabel('x label')  # Add an x-label to the Axes.

ax1.set_ylabel('y label')  # Add a y-label to the Axes.

# title(label, fontdict=None, loc='center', pad=None, **kwargs)
ax1.set_title("Scatter plot \nRandom data", loc='left')  # Add a title to the first axis
ax2.set_title("Bar plot \nFruit data", loc='left')  # Add a title to the second axis

ax1.legend(loc='upper left')  # Add a legends
ax2.legend()  
```
